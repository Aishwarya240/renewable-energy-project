```python
import pandas as pd

energy = pd.read_csv("/home/aishwarya/Downloads/03 modern-renewable-prod.csv")

# Show all column names
print(energy.columns)
```

```python
# Clean column names
energy.columns = energy.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
print(energy.columns)
```

```python
# Sum all energy types to get total production per row
energy['Total_Renewable_TWh'] = (
    energy['Electricity_from_wind_TWh'] +
    energy['Electricity_from_hydro_TWh'] +
    energy['Electricity_from_solar_TWh'] +
    energy['Other_renewables_including_bioenergy_TWh']
)

print(energy[['Entity','Year','Total_Renewable_TWh']].head())
```

```python
# Group by Year to get total global production
total_global = energy.groupby('Year')['Total_Renewable_TWh'].sum().reset_index()
print(total_global.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.lineplot(data=total_global, x='Year', y='Total_Renewable_TWh', marker='o', color='green')
plt.title('Global Renewable Energy Production Over Years')
plt.xlabel('Year')
plt.ylabel('Total Production (TWh)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Check data shape and info
print(energy.shape)
print(energy.info())

# Check missing values
print(energy.isna().sum())

# Check duplicates
print(energy.duplicated().sum())
```

```python
# Drop duplicates if found
energy = energy.drop_duplicates()

# Fill missing numeric values with 0 (or any method you prefer)
numeric_cols = energy.select_dtypes(include='number').columns
energy[numeric_cols] = energy[numeric_cols].fillna(0)
```

```python
energy['Wind_pct'] = (energy['Electricity_from_wind_TWh'] / energy['Total_Renewable_TWh'])*100
energy['Hydro_pct'] = (energy['Electricity_from_hydro_TWh'] / energy['Total_Renewable_TWh'])*100
energy['Solar_pct'] = (energy['Electricity_from_solar_TWh'] / energy['Total_Renewable_TWh'])*100
energy['Bio_pct'] = (energy['Other_renewables_including_bioenergy_TWh'] / energy['Total_Renewable_TWh'])*100

print(energy[['Entity','Year','Wind_pct','Solar_pct','Hydro_pct','Bio_pct']].head())
```

```python
global_energy_type = energy.groupby('Year')[['Electricity_from_wind_TWh',
                                             'Electricity_from_hydro_TWh',
                                             'Electricity_from_solar_TWh',
                                             'Other_renewables_including_bioenergy_TWh']].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=global_energy_type, x='Year', y='Electricity_from_wind_TWh', marker='o', label='Wind')
sns.lineplot(data=global_energy_type, x='Year', y='Electricity_from_hydro_TWh', marker='o', label='Hydro')
sns.lineplot(data=global_energy_type, x='Year', y='Electricity_from_solar_TWh', marker='o', label='Solar')
sns.lineplot(data=global_energy_type, x='Year', y='Other_renewables_including_bioenergy_TWh', marker='o', label='Bioenergy')
plt.title('Global Renewable Energy Production by Type')
plt.ylabel('Production (TWh)')
plt.xlabel('Year')
plt.legend()
plt.show()
```

```python
# Total renewable energy by country
country_total = energy.groupby('Entity')['Total_Renewable_TWh'].sum().sort_values(ascending=False).head(10)
print(country_total)

# Bar plot of top 10 countries
plt.figure(figsize=(10,5))
sns.barplot(x=country_total.index, y=country_total.values, palette='viridis')
plt.xticks(rotation=45)
plt.ylabel('Total Renewable Energy (TWh)')
plt.title('Top 10 Countries by Renewable Energy Production')
plt.show()
```

```python
top_countries = energy.groupby('Entity')[[
    'Electricity_from_wind_TWh',
    'Electricity_from_hydro_TWh',
    'Electricity_from_solar_TWh',
    'Other_renewables_including_bioenergy_TWh'
]].sum().sort_values('Electricity_from_wind_TWh', ascending=False).head(5)

# Stacked bar
top_countries.plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis')
plt.ylabel('Production (TWh)')
plt.title('Renewable Energy Composition for Top 5 Countries')
plt.xticks(rotation=45)
plt.show()
```
